# Z-Score

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as st
```

```python
# Distribution of marks achieved by students in a test. Mean is 60 and standard deviation is 10.
m = 60
s = 10
data = np.random.normal(m, s, 10000)
sns.kdeplot(data, label = "pdf", linewidth=2)
plt.grid()
plt.show()
```

```python
sns.kdeplot(data, label = "pdf", linewidth=2, cumulative=True)
plt.grid()
plt.show()
```

## Problem 1:
What percent of students have marks below 73?

```python
def z_score(mean, std, observatrion):
  z = (observatrion - mean)/std
  return z
```

```python
marks = 73
z = z_score(m, s, marks)
print(f'Z-Score is: {z}')
```

```python
p = (st.norm.cdf(z))*100
print(f'{np.round(p, 3)}% of students have marks less than {marks}')
```

## Problem 2:
What percent of students have marks above 35?

```python
marks = 35
z = z_score(m, s, marks)
print(f'Z-Score is: {z}')
```

```python
p = (st.norm.cdf(z))*100
p = 100 - p
print(f'{np.round(p, 3)}% of students have marks more than {marks}')
```
