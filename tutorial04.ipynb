```python
age_std = np.std(df['age'], ddof = 1)
age_var = np.var(df['age'], ddof = 1)
print(f'Age Standard Deviation: {np.round(age_std, 3)}')
print(f'Age Variance: {np.round(age_var, 3)}')
```
