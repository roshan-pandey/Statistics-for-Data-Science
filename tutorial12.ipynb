# Poisson Distribution

$$ P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!} $$

where $X$ is the random variable representing the number of events in a fixed interval of time or space, and $k$ is a non-negative integer representing the number of events that occur in that interval. The parameter $\lambda$ represents the average rate of events per unit of time or space.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.special import factorial
```

```python
# Set the Poisson distribution parameter Î»
lam = 5  # lambda, the mean of the distribution

# Generate Poisson-distributed data
data = np.random.poisson(lam, 1000)
```

```python
print(f"Mean: {np.mean(data)}")
print(f"Variance: {np.var(data)}")
```

```python
sns.countplot(x = data) # count or frequency plot
plt.grid()
plt.show()
```

```python
x_values = np.arange(data.max()+1) # possible values of the random variable
pmf = np.exp(-lam) * np.power(lam, x_values) / factorial(x_values)  # PMF formula
cdf = np.cumsum(pmf)  # CDF formula
```

```python
plt.plot(pmf)
plt.show()
```

```python
plt.plot(cdf)
plt.show()
```
