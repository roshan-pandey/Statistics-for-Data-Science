# Normal Distribution

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
d1 = np.random.normal(30, 1, 10000)
d2 = np.random.normal(50, 3, 10000)

sns.kdeplot(d1, label = "pdf", linewidth=2)
sns.kdeplot(d2, label = "pdf", linewidth=2)
plt.show()
```

```python
m = 160
s = 5
data = np.random.normal(m, s, 10000)
sns.kdeplot(data, label = "pdf", linewidth=2)

plt.axvline(x = m-s, color = 'b', linewidth = 1)
plt.axvline(x = m+s, color = 'b', linewidth = 1)

plt.axvline(x = m-2*s, color = 'r', linewidth = 1)
plt.axvline(x = m+2*s, color = 'r', linewidth = 1)

plt.axvline(x = m-3*s, color = 'g', linewidth = 1)
plt.axvline(x = m+3*s, color = 'g', linewidth = 1)

plt.show()
```

```python
result = data[(data >= m-s) & (data <= m+s)]
(len(result)/len(data))*100
```

```python
result = data[(data >= m-2*s) & (data <= m+2*s)]
(len(result)/len(data))*100
```

```python
result = data[(data >= m-3*s) & (data <= m+3*s)]
(len(result)/len(data))*100
```
