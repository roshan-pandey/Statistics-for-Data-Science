```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pydataset import data
from scipy.stats import skew
```

```python
# Sample data
data = np.array([2, 3, 3, 4, 8])

# Scale the data by 2
scaled_data = data*2

# Plot the original and scaled data on a single plot
sns.kdeplot(data, label='Original Data')
sns.kdeplot(scaled_data, label='Scaled Data')
plt.legend()
plt.show()
```

```python
# Sample data
data = np.array([2, 3, 3, 4, 8])

# Scale the data by 2
scaled_data = data+2

# Plot the original and scaled data on a single plot
sns.kdeplot(data, label='Original Data')
sns.kdeplot(scaled_data, label='Scaled Data')
plt.legend()
plt.show()
```

```python
scaled_age = df['age']/10
shifted_age = df['age'] + 10
```

```python
def mode(data):
  vals, counts = np.unique(data, return_counts=True)
  mode_idx = np.argwhere(counts==np.max(counts))
  return vals[mode_idx][0][0]
```

```python
def IQR(data):
  q1 = np.percentile(data, 25)
  q3 = np.percentile(data, 75)

  iqr = q3 - q1
  return iqr
```

## Original Data

```python
print(f"Mean: {np.mean(df['age'])}")
print(f"Median: {np.median(df['age'])}")
# print(f"Mode: {mode(df['age'].values)}")
print(f"Variance: {np.var(df['age'])}")
print(f"Range: {df['age'].max() - df['age'].min()}")
print(f"IQR: {IQR(df['age'].values)}")
```

## Stats after scaling

```python
print(f"Mean: {np.mean(scaled_age)}")
print(f"Median: {np.median(scaled_age)}")
# print(f"Mode: {mode(scaled_age.values)}")
print(f"Variance: {np.var(scaled_age)}")
print(f"Range: {scaled_age.max() - scaled_age.min()}")
print(f"IQR: {IQR(scaled_age.values)}")
```

## Stats after shifting

```python
print(f"Mean: {np.mean(shifted_age)}")
print(f"Median: {np.median(shifted_age)}")
# print(f"Mode: {mode(shifted_age.values)}")
print(f"Variance: {np.var(shifted_age)}")
print(f"Range: {shifted_age.max() - shifted_age.min()}")
print(f"IQR: {IQR(shifted_age.values)}")
```
